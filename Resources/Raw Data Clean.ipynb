```python
import pandas as pd
```

```python
raw = pd.read_csv("./Project Test/Raw Data Sample.csv")
raw.head()
```

```python
#remove columns that are not quarter end
for col in raw.columns:
    month_raw = col[-2:]
    #print(month_raw)
    if (month_raw == "01" or month_raw == "02" or month_raw == "04" or month_raw == "05" or month_raw == "07" or 
        month_raw == "08" or month_raw == "10" or month_raw == "11"):
        raw.drop(columns=[col], inplace = True)
    
raw
```

```python
raw_no_blank = raw.dropna(axis = 0, subset = raw.iloc[:,4:len(raw.columns)].columns)
raw_no_blank
```

```python
raw_cleaned = raw_no_blank[(raw_no_blank != 0).all(1)]
raw_cleaned
```
